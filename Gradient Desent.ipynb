```python
import matplotlib.pyplot as plt
import numpy as np

# Define the Qudratic Function and its gradients
def f(x):
  return x**2
def gradient(x):
  return 2*x

# Gradient Desent parameter
x = 10 # Initial guess
learning_rate = 0.1
iterations = 20

# List to store the values for plotting
x_values = [x]
f_values = [f(x)]

# Gradient Desent Loop
for _ in range(iterations):
  grad = gradient(x)
  x = x - learning_rate*grad
  x_values.append(x)
  f_values.append(f(x))

# Plot the fucation and the optimization path
x_range = np.linspace(-10, 10, 400)
plt.plot(x_range, f(x_range), label='$f(x) = x^2$')
plt.scatter(x_values, f_values, color='red')
plt.plot(x_values, f_values, color='red', label='Gradient Desent Path')
plt.xlabel('x')
plt.ylabel('$f(x)$')
plt.title('Gradient Desent Qptimization')
plt.legend()
plt.grid(True)
plt.show()
```

**Gradient Project**

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the Qudratic Function and its gradients
def f(x):
  return (x-3)**2 + 4
def gradient(x):
  return 2*(x-3)

# Gradient Desent parameter
initial_values = [10, -10, 5, -5] # Initial guess
learning_rate = 0.1
iterations = 20

x_range = np.linspace(-10, 10, 400)
plt.plot(x_range, f(x_range), label='$f(x) = (x-3)^2 + 4$')

# Gradient Desent Loop
for x in initial_values:
  x_values = [x]
  f_values = [f(x)]
  for _ in range(iterations):
    grad = gradient(x)
    x = x - learning_rate*grad
    x_values.append(x)
    f_values.append(f(x))
    plt.scatter(x_values, f_values)
plt.plot(x_values, f_values, label=f'Path From Initial x = {x_values[0]}')

# Plot the fucation and the optimization path
plt.xlabel('x')
plt.ylabel('$f(x)$')
plt.title('Gradient Desent Optimization From Different Initial Values')
plt.legend()
plt.grid(True)
plt.show()
```
